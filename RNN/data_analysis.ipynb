```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data/jena_climate_2009_2016.csv')
df.head()
```

```python
print(f"Shape: {df.shape}")
print(f"Date range: {df['Date Time'].min()} to {df['Date Time'].max()}")
print()
print(df.dtypes)
```

```python
df.describe()
```

```python
common_corrupt = [-9999, -999, 999, 9999, -1, 99999]
for val in common_corrupt:
    count = (df == val).sum().sum()
    if count > 0:
        print(f"{val}: found {count} times")
        print((df == val).sum())
        print()
```

```python
print("Humidity > 100%:", (df['rh (%)'] > 100).sum())
print("Wind speed < 0:", (df['wv (m/s)'] < 0).sum())
print("Pressure <= 0:", (df['p (mbar)'] <= 0).sum())
```

```python
print("Missing values:")
print(df.isnull().sum())
print()
print("Corrupted values (-9999):")
print((df == -9999).sum())
```

```python
df['wv (m/s)'] = df['wv (m/s)'].replace(-9999, np.nan)
df['max. wv (m/s)'] = df['max. wv (m/s)'].replace(-9999, np.nan)

print("Corrupted values after cleaning:")
print((df == -9999).sum())
print()
print("Missing values after cleaning:")
print(df[['wv (m/s)', 'max. wv (m/s)']].isnull().sum())
```

```python
df['Date Time'] = pd.to_datetime(df['Date Time'], format='%d.%m.%Y %H:%M:%S')
df.set_index('Date Time', inplace=True)

df.to_csv('data/jena_cleaned.csv')
print("Cleaned data saved.")
print(f"Shape: {df.shape}")
```

```python
monthly_avg = df['T (degC)'].resample('ME').mean()

plt.figure(figsize=(14, 4))
plt.plot(monthly_avg.index, monthly_avg.values)
plt.title('Monthly Average Temperature (2009–2016)')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(), annot=True, fmt='.2f', cmap='coolwarm', center=0)
plt.title('Feature Correlation Matrix')
plt.tight_layout()
plt.show()
```
